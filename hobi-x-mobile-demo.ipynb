```python
# Hobi-X Mobil Uygulama Proje YapÄ±sÄ±
import os
import json

# Proje klasÃ¶r yapÄ±sÄ±nÄ± gÃ¶rselleÅŸtir
def show_directory_structure(path, prefix="", max_depth=3, current_depth=0):
    if current_depth > max_depth:
        return
    
    try:
        items = sorted(os.listdir(path))
        for i, item in enumerate(items):
            if item.startswith('.'):
                continue
            
            item_path = os.path.join(path, item)
            is_last = i == len(items) - 1
            
            current_prefix = "â””â”€â”€ " if is_last else "â”œâ”€â”€ "
            print(f"{prefix}{current_prefix}{item}")
            
            if os.path.isdir(item_path) and current_depth < max_depth:
                next_prefix = prefix + ("    " if is_last else "â”‚   ")
                show_directory_structure(item_path, next_prefix, max_depth, current_depth + 1)
    except PermissionError:
        print(f"{prefix}[Permission Denied]")

print("ðŸ“ Hobi-X Proje YapÄ±sÄ±:")
print("Hobi-X/")
show_directory_structure("/home/garibancoder/MasaÃ¼stÃ¼/Hobi-X", "")
```
